```python
import seaborn as sns
import matplotlib.pyplot as plt

output=df.output
serum_cholestrol=df.serum_cholestrol

sns.swarmplot(x='output',y='serum_cholestrol',data=df)
plt.show()

```

**Higher Cholesterol Levels: Points are more concentrated at higher values(above 200) on the y-axis, suggesting higher cholesterol levels are common among those with heart disease.**

**Higher Cholesterol: If one group (e.g., output = 1) consistently shows higher cholesterol levels compared to the other group (e.g., output = 0), it might indicate a relationship between higher cholesterol levels and the condition represented by output = 1.**

**Overlap: If there is significant overlap in the distribution of cholesterol levels between the groups, it may suggest that cholesterol alone is not a strong differentiator for the output variable.**

**(in both output 0 and 1 persons with and without heart problem , cholestrol level is high,it may suggest that cholesterol alone is not a strong differentiator for the output variable. )**

```python
sns.lmplot(x='age', y='Max_heart_rate', hue='output',col='chest_pain',data=df)
plt.show()
```

**"The lmplot reveals that the maximum heart rate generally decreases with age across all chest pain types.**

**The rate of decline and overall maximum heart rate values differ based on chest pain type and heart disease status.**

Typical angina patients with heart disease show a significant decline in maximum heart rate with age, highlighting the severe impact of heart disease.

In contrast, asymptomatic patients show a similar age-related decline, but those with heart disease still maintain lower maximum heart rates. This analysis underscores the importance of considering both chest pain type and heart disease status when assessing cardiovascular health
